##없는 값 채우기

```python
df.last_review.fillna(method="ffill",inplace=True)
df.first_review.fillna(method="ffill",inplace=True)
df.host_since.fillna(method="ffill",inplace=True)
```

```python
df["bathrooms"] = df['bathrooms'].fillna(round(df["bathrooms"].median()))
```

```python
df["review_scores_rating"] = df["review_scores_rating"].fillna(0)
```

```python
df["bedrooms"] = df['bedrooms'].fillna((df["bedrooms"].median()))
```

```python
df["beds"] = df["beds"].fillna((df['beds'].median()))
```

```python
df.isnull().sum()
```

##정수화

```python
amenities_count = []
for i in df["amenities"]:
    amenities_count.append(len(i))

df["amenities"] = amenities_count
```

```python
categorical_col = []
numerical_col = []
for column in df.columns:

    if df[column].dtypes != "float64" and df[column].dtypes != "int64":
        categorical_col.append(column)
    else:
        numerical_col.append(column)
```

```python
numerical_col
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for col in categorical_col:
    df[col] = le.fit_transform(df[col]) #어떻게 했는지는 모
```

```python
dropcols=["id","name","log_price","description","first_review","host_since","last_review","neighbourhood","thumbnail_url", "zipcode"]
X=df.drop(dropcols, axis=1)
Y=df.log_price
```

##Train Test Split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
from xgboost import XGBRegressor
from sklearn import metrics
xgb = XGBRegressor(objective='reg:squarederror')
xgb.fit(xtrain, ytrain)
ypredict_xgb = xgb.predict(xtest)

mean_square_error_xgb= metrics.mean_squared_error(ytest, ypredict_xgb)
root_mean_square_xgb = np.sqrt(metrics.mean_squared_error(ytest, ypredict_xgb))
r2 = metrics.r2_score(ytest, ypredict_xgb)

print('\nMean Squarred Error of XGBoost: ', mean_square_error_xgb)
print('\nRoot Mean Squarred Error of XGBoost: ', root_mean_square_xgb)
print('\nr2 socre: ', r2)
```
